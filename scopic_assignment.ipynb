```python
from autopep8 import fix_code
import re

def clean_code(code):
    """
    Cleans the given code by performing the following:
    - Removing comments
    - Normalizing variable names
    - Standardizing whitespace
    - Formatting the code consistently

    Args:
        code (str): The raw code string.

    Returns:
        str: The cleaned code string.
    """
    # 1. Remove Single-line Comments (e.g., // or #)
    code = re.sub(r'//.*', '', code)  # For languages like C, C++, Java
    code = re.sub(r'#.*', '', code)   # For languages like Python

    # 2. Remove Multi-line Comments (e.g., /* */)
    code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)

    # 3. Normalize Variable Names
    def normalize_variables(match):
        return f"var{normalize_variables.counter}"
    normalize_variables.counter = 1

    code = re.sub(r'\b[_a-zA-Z][_a-zA-Z0-9]*\b', normalize_variables, code)

    # 4. Standardize Whitespace (Remove Extra Whitespaces)
    code = re.sub(r'[ \t]+', ' ', code)  # Replace multiple spaces/tabs with one space
    code = re.sub(r'\n\n+', '\n', code)  # Reduce multiple newlines to one newline

    # 5. Format Code Consistently
    code = fix_code(code)  # Use autopep8 for Python formatting

    return code
```

```python
prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
Analyze the provided details and predict the likelihood that the Candidate Code is AI-generated. The score ranges from 0 to 1, where:
- 0 indicates no AI characteristics detected.
- 1 indicates a high likelihood of AI characteristics.
Base your analysis on the similarities and differences in syntax, structure, variable naming, and logical flow between the Candidate Code and the AI-Generated Code.

### Input:
{{
    "question": "{}",
    "rules": "{}",
    "examples": "{}",
    "programming_language": "{}",
    "candidate_code": "{}",
    "ai_code": "{}"
}}

### Response
{}"""

EOS_TOKEN = tokenizer.eos_token # Must add EOS_TOKEN
def formatting_prompts_func(row):
    question = row["question"]
    rules = row["rules"]
    examples = row["examples"]
    programming_language = row["programming_language"]
    candidate_code = row["candidate_code"]
    ai_code = row["ai_code"]
    score = row["plagiarism_score"]

    text = prompt.format(question, rules, examples, programming_language, candidate_code, ai_code, score) + EOS_TOKEN
    return text

train_set['input_text'] = train_set.apply(formatting_prompts_func, axis=1)
train_set.head()
```

## Train the model

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "input_text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "/content/drive/MyDrive/scopic assignment/outputs",
    ),
)
```

```python
# Generate AI code
import openai

def generate_ai_code(question):
  # OpenAI API setup (replace YOUR_API_KEY with your actual API key)
  openai.api_key = ""

  # Prompt construction
  prompt = f"""
  Write a program in {question['programming_language']} based on the following details:

  Question:
  {question['question']}

  Rules:
  {', '.join(question['rules'])}

  Examples:
  {question['examples']}

  Ensure the program is efficient and adheres to the rules.
  """

  # Call the OpenAI GPT-4 API to generate the code
  try:
      response = openai.Completion.create(
          engine="gpt-4-turbo",  # GPT-4 engine
          prompt=prompt,
          max_tokens=300,
          temperature=0.5
      )

      # Extract the generated code
      generated_code = response.choices[0].text.strip()
      print("Generated Code:")
      print(generated_code)

  except Exception as e:
      print(f"Error generating code: {str(e)}")
      generated_code = ""

  return generated_code

```

```python
from flask import Flask, request, jsonify
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

app = Flask(__name__)

# Load pre-trained tokenizer and model (replace with your fine-tuned model)
model_name = "path/to/your/fine-tuned-model"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.eval()  # Set model to evaluation mode

# Define a function to preprocess the input and predict the score
def predict_score(ai_code, candidate_code):
    # Prepare the input prompt
    prompt = f"""
    Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

    ### Instruction:
    Analyze the provided details and predict the likelihood that the Candidate Code is AI-generated. The score ranges from 0 to 1, where:
    - 0 indicates no AI characteristics detected.
    - 1 indicates a high likelihood of AI characteristics.
    Base your analysis on the similarities and differences in syntax, structure, variable naming, and logical flow between the Candidate Code and the AI-Generated Code.

    ### Input:
    {{
        "ai_code": "{ai_code}",
        "candidate_code": "{candidate_code}"
    }}

    ### Response:
    """

    # Tokenize the input
    inputs = tokenizer(prompt, return_tensors="pt", max_length=512, truncation=True)

    # Make a prediction
    with torch.no_grad():
        outputs = model(**inputs)
        scores = torch.nn.functional.softmax(outputs.logits, dim=-1)

    # Extract the predicted score (assuming binary classification: 0 = "not AI", 1 = "AI-like")
    ai_score = scores[0, 1].item()  # Probability for the "AI-like" class
    return ai_score

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Parse input data
        data = request.json
        candidate_code = data.get('candidate_code', '')
        question = data.get('question', '')
        ai_code = generate_ai_code(question)


        # Ensure both inputs are provided
        if not ai_code or not candidate_code:
            return jsonify({"error": "Both 'ai_code' and 'candidate_code' are required."}), 400

        # Predict the AI similarity score
        score = predict_score(ai_code, candidate_code)

        # Return the result
        return jsonify({"score": score})

    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == '__main__':
    app.run(debug=True)
```
